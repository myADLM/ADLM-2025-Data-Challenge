# Prepping data for embedding atlas

```python
import pandas as pd
import os
```

```python
atlas_df = pd.read_parquet("app/database/medallions/gold.parquet")[["file_path", "contextual_chunk", "embedding"]]
```

```python
atlas_df["file_path"] = atlas_df["file_path"].apply(lambda x: os.path.dirname(x).split("/")[-1])
```

```python
atlas_df.to_parquet("app/database/medallions/atlasf32.parquet")
```
