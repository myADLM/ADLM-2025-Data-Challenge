```python


def sop_to_es(question, model="o3", DEBUG=0):
    """
    Summarize an SOP document into a JSON object
    """

    lines = ['You are a eloquent, but concise AI assistant who works in the lab. Your expertise is summarizing SOP documents by creating a very concise JSON summary of the document.']
    lines.append('You will be given an SOP document. You should check the document for the Test Name (and purpose of the test), Specimen Type and Container, and Equipment')
    lines.append('You should match the JSON format below exactly. Do not add any additional fields.')
    lines.append('---------')
    lines.append('ORIGINAL SOP Document')
    lines.append(question)
    lines.append('---------')
    lines.append('Return a JSON object with the following fields:')
    lines.append('''        Return JSON with:
        {{
            "Test Name": "name of the test",
            "Specimen Type": "type of specimen",
            "Purpose": "purpose of the test",
            "Container": "type of container",
            "Equipment": "type of equipment"
        }}''')
    lines.append('Make sure to return valid JSON.')
    lines.append('Do not return any additional text')
    lines.append('Here is an example output in the JSON format:')
    lines.append('''        {
                "Test Name": "Glucose, Plasma",
                "Specimen Type": "Plasma",
                "Purpose": "To measure the concentration of glucose in plasma for the diagnosis and management of diabetes mellitus.",
                "Container": "Lithium heparin or sodium fluoride/potassium oxalate tube",
                "Equipment": "Automated biochemistry analyzer (e.g., Roche Cobas c501)"
                }''')
    
    prompt_text = ""
    for msg in lines:
        prompt_text += f"{'System'}: {msg}\n"
    from northwellllm import NorthwellLLM
    llm = NorthwellLLM(model=model)
    llm_response = llm(prompt_text)
    if DEBUG>0:
        print(llm_response)
    if DEBUG>2:
        print(prompt_text)
    return llm_response
```

```python
def es_to_file(es, sop_filename, DEBUG=0):
    """
    Save the es to a file
    """
    import os
    
    # Create es1 directory if it doesn't exist
    os.makedirs('../LabDocs/es1/sop', exist_ok=True)
    
    # Extract just the filename without path and replace extension
    base_filename = os.path.basename(sop_filename).replace('.txt', '_es1.json')
    out_filename = os.path.join('../LabDocs/es1/sop', base_filename)

    with open(out_filename, 'w') as f:
        f.write(es)
    if DEBUG>0:
        print(f"Saved es to {out_filename}")
    return out_filename
```
