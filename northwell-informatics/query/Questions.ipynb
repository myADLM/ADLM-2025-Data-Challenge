# Additional “Indirect” (Diagnosis‐Oriented) Lab Questions

1. Which laboratory tests would you order to evaluate a patient with suspected Cushing’s syndrome?  
2. What combination of serum markers and ratios distinguishes hepatocellular injury from cholestatic liver disease?  
3. How do you use anion gap and osmolar gap calculations to diagnose suspected toxic alcohol ingestion?  
4. Which coagulation and fibrinolysis parameters define disseminated intravascular coagulation (DIC)?  
5. What serologic tests and their cutoff values are used to confirm rheumatoid arthritis?  
6. How can you differentiate nephrotic syndrome from nephritic syndrome based on urine protein and microscopy?  
7. What pattern of troponin and CK-MB rise and fall is diagnostic for acute myocardial infarction?  
8. Which endocrine assays and stimulation tests help confirm primary versus secondary adrenal insufficiency?  
9. What laboratory findings (e.g., LDH, haptoglobin, reticulocyte count) are consistent with hemolytic anemia?  
10. How do you interpret CSF cell count, protein, and glucose to distinguish bacterial meningitis from viral meningitis?  
11. Which electrolyte and osmolality results point toward SIADH versus diabetes insipidus?  
12. What set of tests (e.g., ANA, ANCA, complement levels) is used to evaluate suspected systemic lupus erythematosus (SLE)?  
13. How do peritoneal fluid cell counts and LDH compare to serum values to classify ascites as transudate or exudate?  
14. Which lab values and formulas are used to diagnose and monitor diabetic ketoacidosis (DKA)?  
15. What panels or assays should be ordered to assess for malabsorption syndrome (e.g., celiac disease)?
