```python
sample_questions = [
    "What's the normal range for Hemoglobin A1c?",
    "What kind of blood sample do you need for HbA1c test?",
    "Which tube do you use to collect serum for CMP?",
    "Name 3 tests in the CMP.",
    "How can i check if a patient is pregnant?",
    "Which liver tests are in the CMP?",
    "If HbA1c is 6.8%, what does that mean?",
    "What do you do if QC fails in the lab?",
    "If a blood sample is hemolyzed, which tests might be wrong?",
    "How do you make sure the CD30 stain worked right?",
    "You see hemolysis in a sample for HbA1c. What should you do?",
    "Which tests can be messed up by hemoglobin variants?",
    "CMP shows high K+ and BUN. What might cause that? What's next?",
    "You get an error during HbA1c testing. What steps do you take?",
    "Which sample handling is more critical for HbA1c or CMP?",
    "What problem could ruin both immunostain and blood tests? How to fix?"
]

# sample_questions = [
#         "Which SOPs contain errors??"
# ]

print(f"Testing {len(sample_questions)} sample questions...")
print("=" * 80)

results = []

for i, question in enumerate(sample_questions, 1):
    print(f"\nüîç QUESTION {i}/{len(sample_questions)}:")
    print(f"Q: {question}")
    print("-" * 60)
    
    try:
        sop_fda = SOP_FDA(debug=0)
        res = sop_fda.make_determination(question)  
        print(f"{res}")
    except Exception as e:
        res = f"‚ùå ERROR: {e}"
        print(res)
    
    results.append({"question": question, "determination": res})
    print("=" * 80)

# Output concise JSON
print("\nConcise JSON output:")
print(json.dumps(results, indent=2))

# Save JSON to file with timestamp
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_path = f"/labs/jupyter/ai_challenge/IsaiahBlanks/jsons/results_{timestamp}.json"
with open(output_path, "w") as f:
    json.dump(results, f, indent=2)
print(f"JSON saved to {output_path}")
```
